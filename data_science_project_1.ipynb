# 1. เพศเกี่ยวข้องกับแนว Anime ยังไงบ้าง?

### 1.1 Pre-process
นำ anime_id ทั้งหมดจาก animelists_cleaned ไป merge เข้ากับ anime_cleaned เพื่อเอา genre จากนั้นนำ genre ซึ่งเป็น string ไป split เพื่อให้กลายเป็น list จากนั้นจึง loop นับว่า user เพศไหนดู genre ไหนบ้าง (ทำอย่างงี้กับอายุเช่นเดียวกัน)


```
df_temp = animelists_cleaned.loc[:, ['username', 'anime_id']]
anime_id_temp = pd.DataFrame(data = df_temp['anime_id'].unique(), columns=['anime_id'])
anime_id = anime_id_temp.assign(genre = anime_id_temp.merge(anime_cleaned, on='anime_id')['genre'])

del anime_id_temp

anime_id.loc[:, 'genre'] = anime_id['genre'].str.split(',')
anime_id
```

```
df = df_temp.groupby(by='username').agg(['unique'])
del df_temp

df.head()
```

```
gender = users_cleaned.loc[:, ['username', 'gender']].set_index('username')
gender.head()
```

```
import datetime 

def to_age(born):
    today = datetime.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

users_cleaned['birth_date'] = pd.to_datetime(users_cleaned['birth_date'])
age = pd.DataFrame(data={"username": users_cleaned['username'], "age": users_cleaned['birth_date'].apply(lambda x: to_age(x))})
age
```

```
age = age.set_index('username')
age.head()
```

```
cnt = 0
counter = {"Male": {}, "Female": {}, "AGE": {}}
for i, row in df.iterrows():
    username = i
    anime_list_ids = row['anime_id']['unique']
    _gender = gender.loc[username]['gender']
    _age = age.loc[username]['age']
    if _age not in counter["AGE"]:
      counter["AGE"][_age] = {}
    for id in anime_list_ids:
      try:
        for cate in anime_id.loc[id]['genre']:
          cate = cate.strip()
          # count by age
          if cate in counter["AGE"][_age]:
            counter["AGE"][_age][cate] = counter["AGE"][_age][cate] + 1
          else:
            counter["AGE"][_age][cate] = 1
          
          # count by gender
          if cate in counter[_gender]:
            counter[_gender][cate] = counter[_gender][cate] + 1
          else:
            counter[_gender][cate] = 1
      except Exception:
        pass
    cnt = cnt + 1
    print(str(cnt) + ' ' + username)
 
print(counter)
```

### 1.2 Summary

#### 1.2.1 Percent การดู Anime ในแต่ละประเภทแบ่งตามเพศ

```
male = pd.DataFrame.from_dict(counter["Male"], orient='index').reset_index()
male.columns = ['genre', 'total']
male = male.sort_values(by='total', ascending=0)
male.head()
```

```
female = pd.DataFrame.from_dict(counter["Female"], orient='index').reset_index()
female.columns = ['genre', 'total']
female = female.sort_values(by='total', ascending=0)
female.head()
```

```
f,ax = plt.subplots(figsize=(30,15))
sns.barplot(data=male, x='genre', y='total', label='Male', color='r', alpha = 0.7)
sns.barplot(data=female, x='genre', y='total', label='Female', color='b', alpha = 0.4)

ax.set(xlabel='Genres', ylabel='Total', title='Percentage of Anime Genres According to Gender')
ax.legend(loc='lower right',frameon= True)
plt.show()
```

#### 1.2.2 อัตราส่วนการดู Anime ในแต่ละเพศเทียบกับเพศเดียวกัน

```
male['total'] = male['total'] / male['total'].sum()
female['total'] = female['total'] / female['total'].sum()

df2 = male.merge(female, on='genre')
df2.columns = ['genre', 'male', 'female']

df2.head()
```

```
_df2 = df2.melt(id_vars=['genre'], value_vars=['male', 'female'])
f,ax = plt.subplots(figsize=(30,15))
sns.barplot(data=_df2, x='genre', y='value', alpha = 0.8, hue='variable')

ax.set(xlabel='Genres', ylabel='Total', title='Ratio of Anime Genres According to Gender')
ax.legend(loc='lower right',frameon= True)
plt.show()
```

# 3.ความเกี่ยวข้องระหว่างประเทศกับ Genres ของ Anime

### 3.1 Pre-process
clean ข้อมูล location ใน users_cleaned โดยนำ dataset ISO มาเป็นตัวเปรียบเทียบเพื่อปรับข้อมูล location ให้เป็นกลายเป็นประเทศที่อยู่และตัดข้อมูลที่ไม่จำเป็นออก

#### คำอธิบาย Dataset ISO3166.csv
- Country = ชื่อประเทศ (string)
- Alpha-2 code = รหัสประเทศแบบ (string)
- Alpha-3 code = รหัสประเทศแบบ (string)
- Numeric = รหัสประเทศแบบตัวเลข (integer)

```
iso = pd.read_csv("ISO3166.csv")
```

```
iso.head(5)
```

```
country = iso.Country
code = iso['Alpha-2 code'].astype(str)
threecode = iso['Alpha-3 code']
```

```
before = users_cleaned.location.astype(str)
after = []
for i in range(before.size):
    check = False
    for j in range(country.size):
        if (code[j] in before[i]) or (country[j] in before[i]) or (threecode[j] in before[i]):
            idx = j
            check = True
            break
    if check:
      after.append(country[idx])
    else:
      after.append("unknown")
```

```
users_cleaned.location = after
```

#### 3.1.2 รวมตาราง animelist_cleaned และตาราง users_cleaned เป็นตาราง users_location 
ตาราง users_cleaned จะเน้นเฉพาะข้อมูล location เนื่องจากขนาดของข้อมูลค่อนข้างใหญ่

```
users_location = animelists_cleaned.loc[:,('username','anime_id')].merge(users_cleaned.loc[:,('username','location')],left_on='username',right_on='username',how='left')

# clear memory
del animelists_cleaned

users_location
```

#### 3.1.3 สร้างตาราง genre_anime แบ่งแยกข้อมูล genre ในตาราง anime_cleaned

```
genre_anime = anime_cleaned.genre.str.split(',', expand=True).stack().to_frame(name='genre').droplevel(1)
genre_anime.genre = genre_anime.genre.replace({' ':''},regex=True)
print("DONE")
```

```
genre_anime = genre_anime.reset_index()
genre_anime.columns = ['id','genre']
print("DONE")
```

```
genre_anime_temp = anime_cleaned.anime_id.to_frame().merge(genre_anime,how='left',left_on=anime_cleaned.index,right_on='id')
genre_anime = genre_anime_temp.drop('id',axis=1)
del genre_anime_temp
print("DONE")
```

#### 3.1.4 สร้างตาราง summary_loc โดยการรวมตาราง genre_anime และตาราง users_location
ข้อมูลใน users_location จะเลือกเฉพาะข้อมูลที่รู้ location

```
df_temp = users_location.loc[users_location.location!='unknown']
del users_location

summary_loc_temp = df_temp.merge(genre_anime, how='left')
del df_temp

summary_loc = summary_loc_temp.groupby(['location','genre']).size().reset_index()
del summary_loc_temp

summary_loc.columns = ['location','genre','count']
print("DONE")
```

# 5. Anime ในชุดข้อมูลมี rating อะไรบ้างอย่างละกี่ % และแนวโน้มของ rating ที่มีมากที่สุด

```
anime_rate = anime_cleaned
anime_rate[['rating','rate']] = anime_rate.rating.str.split("- ",expand=True)
anime_rate.loc[:,['rating','rate']]
```
